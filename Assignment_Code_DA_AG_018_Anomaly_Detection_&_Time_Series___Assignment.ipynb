**Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.**

**Answer:**

**Anomaly Detection** is a data analysis technique used to identify data points, patterns, or observations that deviate significantly from the normal or expected behavior of a system. These unusual patterns are called **anomalies** or **outliers**. Anomaly detection is widely used in applications such as fraud detection, network security, fault detection, and healthcare monitoring.

There are three main types of anomalies:

---

### 1. Point Anomaly

A **point anomaly** occurs when a single data point is significantly different from the rest of the data.

**Example:**

* In a bank transaction system, if a customer who usually spends ₹1,000–₹2,000 suddenly makes a transaction of ₹1,00,000, this transaction is a point anomaly and may indicate fraud.
* A temperature sensor normally records values around 25°C, but suddenly records 90°C—this is a point anomaly.

---

### 2. Contextual Anomaly

A **contextual anomaly** occurs when a data point is anomalous only in a specific context (such as time, location, or season).

**Example:**

* A temperature of 30°C is normal during summer but would be considered anomalous during winter.
* High electricity usage during the daytime may be normal, but the same usage at midnight could be a contextual anomaly.

---

### 3. Collective Anomaly

A **collective anomaly** occurs when a group of related data points is anomalous, even if individual points appear normal on their own.

**Example:**

* A sudden sequence of small but continuous network requests may indicate a cyber-attack (DDoS attack), even though each request individually looks normal.
* A series of slight increases in heart rate readings over time may collectively indicate a health problem.

---

**Conclusion:**
Anomaly detection helps in identifying unusual patterns in data. Point anomalies focus on individual data points, contextual anomalies depend on the surrounding context, and collective anomalies involve a group of data points that together indicate abnormal behavior.



**Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.**

**Answer:**

Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) are popular techniques used for anomaly (outlier) detection. They differ in their working approach and the types of problems they are best suited for.

---

### 1. Isolation Forest

**Approach:**
Isolation Forest is a tree-based anomaly detection algorithm. It works by randomly selecting a feature and then randomly selecting a split value to isolate data points. Anomalies are isolated faster because they are few and very different from normal data points.

**Key Idea:**
Anomalies require fewer splits to be isolated in a tree structure.

**Suitable Use Cases:**

* Large datasets with high dimensionality
* Fraud detection (banking, insurance)
* Network intrusion detection
* Situations where labeled data is not available

---

### 2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

**Approach:**
DBSCAN is a density-based clustering algorithm. It groups closely packed data points into clusters and labels points that lie in low-density regions as noise (outliers).

**Key Idea:**
Normal data points belong to dense regions, while anomalies lie in sparse regions.

**Suitable Use Cases:**

* Spatial and geographical data
* Datasets with clusters of arbitrary shapes
* Detecting noise and outliers in clustering tasks
* Small to medium-sized datasets

---

### 3. Local Outlier Factor (LOF)

**Approach:**
LOF measures the local density of a data point and compares it with the density of its neighbors. A point is considered an outlier if it has significantly lower density than its surrounding points.

**Key Idea:**
Anomalies are points with much lower local density compared to neighbors.

**Suitable Use Cases:**

* Detecting local anomalies in datasets with varying densities
* Situations where global methods fail to detect subtle outliers
* Fraud detection and intrusion detection with local patterns

---

### Comparison Summary

| Algorithm            | Approach                 | Best Use Cases                 |
| -------------------- | ------------------------ | ------------------------------ |
| Isolation Forest     | Tree-based isolation     | Large, high-dimensional data   |
| DBSCAN               | Density-based clustering | Spatial data, noise detection  |
| Local Outlier Factor | Local density comparison | Local and contextual anomalies |

---

**Conclusion:**
Isolation Forest is efficient for large and high-dimensional datasets, DBSCAN is effective for density-based clustering and noise detection, and LOF is best suited for identifying local outliers where data density varies.



**Question 3: What are the key components of a Time Series? Explain each with one example.**

**Answer:**

A **Time Series** is a sequence of data points collected or recorded at regular time intervals. Time series analysis helps in understanding patterns, trends, and behaviors over time. The main components of a time series are **Trend, Seasonal, Cyclical, and Irregular (Random)** components.

---

### 1. Trend Component

The **trend** represents the long-term movement or direction of the data over time. It shows whether the data is increasing, decreasing, or remaining constant.

**Example:**

* A company’s annual sales increasing steadily over several years due to business growth.

---

### 2. Seasonal Component

The **seasonal** component shows regular, repeating patterns that occur at fixed intervals (daily, monthly, quarterly, yearly) due to seasonal factors.

**Example:**

* Ice cream sales increase every summer and decrease during winter each year.

---

### 3. Cyclical Component

The **cyclical** component represents fluctuations that occur over long periods, but not at fixed intervals. These are usually related to economic or business cycles.

**Example:**

* Changes in unemployment rates during economic booms and recessions.

---

### 4. Irregular (Random) Component

The **irregular** or **random** component consists of unpredictable and unexpected variations caused by random events.

**Example:**

* A sudden drop in airline bookings due to an unexpected natural disaster or pandemic.

---

**Conclusion:**
A time series is made up of trend, seasonal, cyclical, and irregular components. Understanding these components helps in better forecasting and decision-making.


**Question 4: Define Stationary in Time Series. How can you test and transform a non-stationary series into a stationary one?**

**Answer:**

### Stationary Time Series

A **stationary time series** is one whose statistical properties remain constant over time. In a stationary series:

* The **mean** is constant
* The **variance** is constant
* The **autocovariance (correlation)** depends only on the lag, not on time

Stationarity is important because many time series models (like ARIMA) assume the data is stationary.

**Example:**

* Daily temperature deviations around an average value without a long-term upward or downward trend.

---

### Testing Stationarity

#### 1. Visual Inspection

* Plot the time series.
* If the mean or variance changes over time, the series is likely non-stationary.

**Example:**

* A sales graph showing a continuous upward trend indicates non-stationarity.

---

#### 2. Augmented Dickey–Fuller (ADF) Test

* A statistical test used to check stationarity.
* **Null hypothesis (H₀):** The series is non-stationary.
* If the **p-value < 0.05**, we reject H₀ and conclude the series is stationary.

---

#### 3. KPSS Test

* Another statistical test for stationarity.
* **Null hypothesis (H₀):** The series is stationary.
* If the **p-value < 0.05**, the series is non-stationary.

---

### Transforming a Non-Stationary Series into a Stationary One

#### 1. Differencing

* Subtract the previous observation from the current one.
* Removes trend and seasonality.

**Example:**
[
Y'*t = Y_t - Y*{t-1}
]

---

#### 2. Logarithmic or Power Transformation

* Stabilizes variance when data grows exponentially.

**Example:**

* Apply log transformation to rapidly increasing sales data.

---

#### 3. Detrending

* Remove the trend component using regression or smoothing methods.

**Example:**

* Subtracting a fitted linear trend from stock price data.

---

#### 4. Seasonal Adjustment

* Remove seasonal patterns using seasonal differencing or decomposition.

**Example:**
[
Y'*t = Y_t - Y*{t-12} \quad \text{(for monthly data)}
]

---

**Conclusion:**
A stationary time series has constant statistical properties over time. Stationarity can be tested using visual inspection, ADF test, and KPSS test. Non-stationary series can be transformed into stationary ones using differencing, transformations, detrending, and seasonal adjustment, enabling accurate time series modeling and forecasting.


**Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.
(Dataset reference: NYC Taxi Fare Data, AirPassengers Dataset)**

**Answer:**

AR, MA, ARIMA, SARIMA, and SARIMAX are time series models used for forecasting. They differ in structure (how they use past data) and in their suitability for different types of datasets such as **NYC Taxi Fare Data** and the **AirPassengers Dataset**.

---

### 1. AR (AutoRegressive) Model

**Structure:**

* Uses past values of the same variable for prediction.
* Represented as **AR(p)**, where *p* is the number of lagged observations.

**Application:**

* Works only on **stationary** time series.
* No seasonality handling.

**Dataset Example:**

* **NYC Taxi Fare Data:**
  If average taxi fares (after removing trend) depend mainly on fares from previous days, an AR model can be applied.

---

### 2. MA (Moving Average) Model

**Structure:**

* Uses past error terms (shocks) instead of past values.
* Represented as **MA(q)**, where *q* is the number of lagged error terms.

**Application:**

* Suitable for stationary data with short-term random fluctuations.

**Dataset Example:**

* **NYC Taxi Fare Data:**
  Useful when fare changes are influenced by sudden random events like traffic jams or weather-related noise.

---

### 3. ARIMA (AutoRegressive Integrated Moving Average)

**Structure:**

* Combines AR and MA with **differencing (I)** to make non-stationary data stationary.
* Represented as **ARIMA(p, d, q)**.

**Application:**

* Used for **non-stationary** data without seasonality.

**Dataset Example:**

* **NYC Taxi Fare Data:**
  Taxi fares often show trends over time. ARIMA can be used after differencing to forecast future fares.

---

### 4. SARIMA (Seasonal ARIMA)

**Structure:**

* Extends ARIMA by adding **seasonal components**.
* Represented as **SARIMA(p, d, q)(P, D, Q)_s**, where *s* is the seasonal period.

**Application:**

* Best for data with **clear seasonal patterns**.

**Dataset Example:**

* **AirPassengers Dataset:**
  Passenger numbers increase every year and show strong monthly seasonality. SARIMA effectively captures both trend and seasonality.

---

### 5. SARIMAX (Seasonal ARIMA with eXogenous Variables)

**Structure:**

* SARIMA model with additional **external (exogenous) variables**.

**Application:**

* Suitable when the target variable is affected by outside factors.

**Dataset Example:**

* **NYC Taxi Fare Data:**
  Taxi fares influenced by fuel prices, demand, weather, or traffic conditions can be modeled using SARIMAX.
* **AirPassengers Dataset:**
  Can include variables like ticket prices, holidays, or economic indicators.

---

### Comparative Summary Table

| Model   | Structure               | Seasonality | External Variables | Suitable Dataset             |
| ------- | ----------------------- | ----------- | ------------------ | ---------------------------- |
| AR      | Past values             | No          | No                 | NYC Taxi Fare (stationary)   |
| MA      | Past errors             | No          | No                 | NYC Taxi Fare (noise-based)  |
| ARIMA   | AR + MA + differencing  | No          | No                 | NYC Taxi Fare (trend)        |
| SARIMA  | ARIMA + seasonal terms  | Yes         | No                 | AirPassengers                |
| SARIMAX | SARIMA + exogenous vars | Yes         | Yes                | NYC Taxi Fare, AirPassengers |

---

**Conclusion:**

* **NYC Taxi Fare Data** is best suited for AR, MA, ARIMA, or SARIMAX depending on trend and external factors.
* **AirPassengers Dataset** is a classic example where **SARIMA** (or SARIMAX) is most effective due to strong seasonality.

## **Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.**

### **Answer:**

Isolation Forest is an unsupervised anomaly detection algorithm that isolates anomalies by randomly partitioning the data. Anomalies are isolated faster because they are few and different from normal observations.

Here, Isolation Forest is applied to a numerical **NYC Taxi Fare–like dataset** using **fare amount** and **trip distance**.

---

### **Python Code:**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Create a synthetic NYC Taxi Fare dataset
np.random.seed(0)
fare_amount = np.random.normal(15, 4, 500)
trip_distance = np.random.normal(3, 1, 500)

# Inject anomalies
fare_amount[:25] += np.random.normal(35, 5, 25)
trip_distance[:25] += np.random.normal(10, 2, 25)

df = pd.DataFrame({
    "fare_amount": fare_amount,
    "trip_distance": trip_distance
})

# Apply Isolation Forest
model = IsolationForest(contamination=0.05, random_state=42)
df["anomaly"] = model.fit_predict(df)

# Separate normal points and anomalies
normal = df[df["anomaly"] == 1]
anomalies = df[df["anomaly"] == -1]

# 2D Scatter Plot
plt.figure()
plt.scatter(normal["fare_amount"], normal["trip_distance"])
plt.scatter(anomalies["fare_amount"], anomalies["trip_distance"])
plt.xlabel("Fare Amount")
plt.ylabel("Trip Distance")
plt.title("Isolation Forest – Anomaly Detection on NYC Taxi Fare Data")
plt.show()
```

---

### **Output Explanation:**

* The **scatter plot** displays fare amount vs trip distance.
* **Normal data points** form dense clusters.
* **Anomalies** appear as isolated points with unusually high fare amounts and long trip distances.
* These anomalous trips may indicate:

  * Data errors
  * Fraudulent rides
  * Unusual traffic or routing behavior

---

### **Conclusion:**

Isolation Forest successfully detects anomalous taxi trips by isolating rare and extreme fare–distance combinations. It is highly effective for large, numerical datasets like NYC Taxi Fare data.

## **Question 9: Apply Local Outlier Factor (LOF) on a numerical dataset to detect anomalies and visualize them using matplotlib.**

### **Answer:**

**Local Outlier Factor (LOF)** is an unsupervised anomaly detection algorithm that identifies anomalies by comparing the **local density** of a data point with the densities of its neighbors. Points with significantly lower local density than their neighbors are considered anomalies.

Below, LOF is applied to a **numerical dataset** (synthetic 2D data), and anomalies are visualized using **matplotlib**.

---

### **Python Code (Code Box):**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Step 1: Create a numerical dataset
np.random.seed(42)
x_normal = np.random.normal(0, 1, 300)
y_normal = np.random.normal(0, 1, 300)

# Add anomalies
x_anomaly = np.random.uniform(6, 8, 20)
y_anomaly = np.random.uniform(6, 8, 20)

x = np.concatenate([x_normal, x_anomaly])
y = np.concatenate([y_normal, y_anomaly])

df = pd.DataFrame({"X": x, "Y": y})

# Step 2: Apply Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.06)
df["anomaly"] = lof.fit_predict(df[["X", "Y"]])

# Step 3: Separate normal points and anomalies
normal = df[df["anomaly"] == 1]
anomalies = df[df["anomaly"] == -1]

# Step 4: Visualize using 2D scatter plot
plt.figure()
plt.scatter(normal["X"], normal["Y"])
plt.scatter(anomalies["X"], anomalies["Y"])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.show()
```

---

### **Output Explanation:**

* The scatter plot shows two groups:

  * **Normal data points** clustered around the center
  * **Anomalies** appearing far from dense regions
* LOF detects anomalies based on **local density differences**, not just global distance.
* Points marked as anomalies are isolated compared to their neighbors.

---

### **Conclusion:**

Local Outlier Factor effectively detects **local anomalies** in numerical datasets. It is especially useful when data has regions of varying density, where global methods may fail to identify subtle outliers.
